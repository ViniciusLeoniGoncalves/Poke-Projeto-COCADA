```julia
using HTTP
using JSON
#entendendo como pegar os nomes dos pokemons dentro da cadeia evolutiva
function get_evolution_names(pokemon_ID::Int)
    # Obtém os dados da espécie do Pokémon
    url = "https://pokeapi.co/api/v2/pokemon-species/$pokemon_ID/"
    response = HTTP.get(url)
    data = JSON.parse(String(response.body))
    
    # URL da cadeia de evolução
    evolution_chain_url = data["evolution_chain"]["url"]
    response_chain = HTTP.get(evolution_chain_url)
    data_chain = JSON.parse(String(response_chain.body))
    chain = data_chain["chain"]

    # Função auxiliar para percorrer a cadeia de evolução de forma recursiva
    function parse_chain(chain_node)
        names = [chain_node["species"]["name"]]
        for evolution in chain_node["evolves_to"]
            append!(names, parse_chain(evolution))
        end
        return names
    end

    # Processa a cadeia de evolução
    return parse_chain(chain)
end
```

## **Capturando os Status dos Pokémon via PokeAPI**  
***  
Neste universo de Pokémon, cada Pokémon possui 6 **status base**:   , que serão interpretados como vetores: $\begin{bmatrix} HP. \\ Atk. \\ Def. \\ Sp. Atk. \\ Sp. Def. \\ Speed\end{bmatrix} $. 

Para isso, criamos duas estruturas principais:  

- **`NAMES_VECTOR`**: Um vetor que armazena os nomes dos 151 Pokémon originais, ordenados pelo número do Pokédex.  
- **`STATS_MATRIX`**: Uma matriz onde cada coluna representa o vetor de status de um Pokémon, também ordenado pelo número do Pokédex.  

Com essa estrutura, acessar a posição $i$ do **`NAMES_VECTOR`** retorna o nome do Pokémon cujo vetor de status está na $i$-ésima coluna da **`STATS_MATRIX`**.  

### **A coleta de Dados via PokeAPI**  
Como a PokeAPI permite acessar dados de cada Pokémon usando seu número do Pokédex, e esse número varia de 1 a 151 para a primeira geração, a consulta se torna iterável, facilitando assim a criação das nossas estruturas principais.  

---

```julia
using HTTP, JSON


#Função para obter informações de um Pokémon por ID
function get_pokemon_data(id::Int)
    url = "https://pokeapi.co/api/v2/pokemon/$id"
    response = HTTP.get(url)
    data = JSON.parse(String(response.body))
    return data
end

function find_id_by_name(pokemon_name)
    url = "https://pokeapi.co/api/v2/pokemon/$pokemon_name"
    response = HTTP.get(url)
    data = JSON.parse(String(response.body))
    return data["id"]
end

function find_name_by_id(pokemon_id)
    data = get_pokemon_data(pokemon_id)
    return data["name"]
end


#função pra saber se o pokemon não é o mais evoluído de sua cadeia
function has_next_evo(pokemon_name)
    # Obtém os dados da espécie do Pokémon
    pokemon_ID = find_id_by_name(pokemon_name)
    url = "https://pokeapi.co/api/v2/pokemon-species/$pokemon_ID/"
    response = HTTP.get(url)
    data = JSON.parse(String(response.body))
    
    pokemon_name = data["name"]

    # Pega a url de cadeia evolutiva
    evolution_chain_url = data["evolution_chain"]["url"]
    chain_response = HTTP.get(evolution_chain_url)
    chain_data = JSON.parse(String(chain_response.body))
    chain = chain_data["chain"]

    # Função recursiva para encontrar o Pokémon e verificar se ele evolui
    function find_and_check(chain_node)
        if chain_node["species"]["name"] == pokemon_name
            return !isempty(chain_node["evolves_to"])
        end

        for evo in chain_node["evolves_to"]
            result = find_and_check(evo)
            if result !== nothing
                return result
            end
        end

        return nothing  # Não encontrado
    end

    return find_and_check(chain) === true
end

#println("Cadeia de evolução: ")
#println(has_next_evo("butterfree")) # Teste com o Pokémon Caterpie e butterfree (ID = 10 e 12)


function create_pokemon_stats_matrix(qt_pkmn::Int)
    matriz_stats = zeros(6, 0)  #Matriz com 6 linhas e 0 colunas
    vetor_nomes = String[] #Matriz com os nomes dos Pokémons
    
    
    for id in 1:qt_pkmn
        data = get_pokemon_data(id)
        nova_coluna = Int[]
        for h in data["stats"]
            push!(nova_coluna, h["base_stat"]) 
        end

        push!(vetor_nomes, data["name"]) 
        
        matriz_stats = hcat(matriz_stats, nova_coluna)
    end

    return matriz_stats , vetor_nomes
end
function deleta_coluna(A,j)
    A_new = hcat(A[:, 1:j-1], A[:, j+1:end]) #separa A em duas matrizes sem a coluna J e une de novo
    return A_new
end

function deleta_linha_vector(V,l) #Remove o elemento na linha l no vetor V
    V_new = copy(V)
    deleteat!(V_new, l)
    return V_new
end

function remove_pre_evos(matrizStats, vetorNomes)
    id=1
    while id <= size(matrizStats,2) #olhar todas as colunas de matriz_stats
        temproximaEvo=has_next_evo(vetorNomes[id])
        if temproximaEvo == true
            matrizStats = deleta_coluna(matrizStats,id)
            vetorNomes = deleta_linha_vector(vetorNomes,id)
        else
            id+=1 
        end
    end 
    return matrizStats, vetorNomes
end

```

## **Redução de dimensionalidade (PCA)**  

Como já falei anteriormente, **o vetor de status possui 6 dimensões, é impossível visualizarmos assim**. Para isso, usaremos a técnica de análise de componentes principais (PCA) para pode fazer uma redução na dimensão dos nossos dados e poder representar, com boa fidelidade, nossos dados em 1, 2 e 3 dimensões. Para executar o PCA fazemos:



Tendo nossa matriz de dados $A$, onde cada coluna é um pokémon, computaremos a matriz de relação (covariância) como $C = A*A^t$, depois acharemos os três autovetores associados aos três maiores autovalores de $C$.
Com os 3 autovetores de C, projetamos todos os vetores de pokémons em cada um deles e separemos os resultados em vetores $p_1$, $p_2$ e $p_3$ 

Por fim, para obter a matriz dos dados em $\mathbb{R}^1$, $\mathbb{R}^2$ ou $\mathbb{R}^3$ temos que fazer a operação: $$V^t*A$$ onde $V$ é a matriz de autovetores (a dimensão dependerá de quantos autovetores usarmos nela) E $A$ é a matriz original. (isso é o mesmo que projetar as colunas de $A$ nos autovetores de $C$)

Agora vamos escrever as funções necessárias para realizar os cálculos com nossa matriz de status pokemon.

```julia
using Statistics
using LinearAlgebra
using Random
Random.seed!(42)
#retorna uma matriz com a quantidade desejada de autovetores da matriz de covariancia de A e uma matruz diagonal com seus autovalores

function centralizar(A)
  media_col = mean(A, dims=2)    # vetor média (n x 1)
  A_cent = A .- media_col
  return A_cent
end

function PCA(A, qt_autovetores)
  C = A*A'
  Clinhas = size(C,1)
  V = zeros(Clinhas, 0) #matriz pra guardar autovetores

  D = zeros(qt_autovetores,0) #matriz diagonal de autovalores

  for i in 1:qt_autovetores
    j=0
    x=normalize(randn(size(C,1)))#chute normalizado
    while j<1000 #metodo da potencia
        x=C*x
        x=normalize(x)
        j+=1
    end
    #x é autovetor associado ao maior autovalor de A

    V = hcat(V,x)
    autoval=(x'*C*x) #x esta normalizado
    d=zeros(qt_autovetores,1)
    d[i,1] = autoval
    D = hcat(D,d)
    C = C - autoval*x*x' #remover influencia do autovalor associado a x que já pegamos, pra que possamos pegar outro autovalor (deflação)

  end

  return V, D
end


function aproxima_A(A, V) #retorna a matriz de posto k que mais se aproxima de A (V e D precisam vir do PCA)
  P = V'*A 
    return V*P
end

function reduz_A(A, V) #reduz a matriz A_n.m para uma outra na dimensão dada R_dim.m (V e D precisam vir do PCA)
  P = V'*A
    return P
end

function erro_aprox(A, Aaprox)
    erro = 0.0
    for j in 1:size(A, 2)
        erro += norm(A[:, j] - Aaprox[:, j])^2 #soma a norma da diferença entre os vetores reais e aproximados de duas matrizeas
    end
    return erro
end
```

## **Clusterização de grupos**  

Depois de reduzir a dimensionalidade dos dados com o PCA, temos representações mais simples e visuais dos pokémons, o que nos permite buscar padrões, grupos e similaridades entre eles. Para isso, vamos utilizar a **técnica de clusterização de $k$-means**

A ideia é simples:

- Escolhemos um número de grupos $k$ (por exemplo, 3).

- Escolhemos $k$ centróides (pokémons) aleatórios.

- Atribuímos cada pokémon ao centróide mais próximo (formando os clusters).

- Recalculamos os centróides com base nos membros de cada grupo.

- Repetimos os passos 3 e 4 até os centróides estabilizarem (ou atingir um número máximo de iterações).

Ao final, teremos grupos de pokémons com características similares entre si, o que nos ajudará a entender quais conjuntos de pokémons podem ser considerados parecidos e identificar padrões de combate como **Blitzkrieg** que queremos.

Dentro desses clusters, se acharmos o de **Blitzkrieg**, vamos fazer a análise artesanal de vantagem de tipo para cada pokémon da Cynthia.

Também faremos uma análise de erro por quantidade de cluster, para entendermos quantos Clusters são suficientes, dado a quantidade de pokémons que estamos usando

```julia
using Statistics
using LinearAlgebra
using Clustering
using Random
Random.seed!(42) #pra que a clusterização sempre dê a mesma


function listar_pokemons_por_cluster(clusters, nomes, k)
    for i in 1:k
        println("Cluster $i:")
        idxs = findall(clusters .== i)
        for j in idxs
            print(" ", nomes[j], ",")
        end
        println()
    end
end

A_reduzido = A2
k=9

#realiza a clusterização em 2 grupos (como não vimos esse metodo em aula, estou usando como 'caixa preta' conforme permitido pelo professor)
resultado = kmeans(A_reduzido, k)

#extrai clusters (vetor com índices 1, 2, ..., k)
clusters = resultado.assignments
#printar quais pokémons estão nos clusters
listar_pokemons_por_cluster(clusters,NAMES_VECTOR,k)

#separar os pontos por cluster pra plotar com cores diferentes
cores = [:red, :blue, :green, :orange, :purple, :cyan, :brown, :black, :pink, :silver, :gold] 
Plots.plot()

for i in 1:k
    #pegamos os índices dos pokémons que caíram no cluster i
    idx = findall(clusters .== i)

    #plotamos os pontos desse cluster com uma cor específica
    Plots.scatter!(A_reduzido[1, idx], A_reduzido[2, idx], label="Cluster $i", color=cores[i])
end

#centróides
scatter!(resultado.centers[1, :], resultado.centers[2, :], markershape=:circle, markersize=10, label="Centroids", color=:YELLOW)

#gráfico final
display(plot!(xlabel="PCA1", ylabel="PCA2", title="Clusters de Pokémons"))
```
